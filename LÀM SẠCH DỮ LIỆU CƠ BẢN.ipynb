##Phần 2: Khám phá và Làm sạch dữ liệu


## Kiểm tra xem mỗi cột trong tập dữ liệu có bao nhiêu giá trị bị thiếu (NaN/Null).


```
# Đếm số giá trị bị thiếu ở mỗi cột
print(df.isnull().sum())
```

#Tiến hành xử lý giá trị thiếu ở cột Discount : Thay thế tất cả các giá trị bị thiếu bằng 0 (giả định đơn hàng không áp dụng giảm giá).

```
df['Discount'] = df['Discount'].fillna(0)
print("Số giá trị thiếu trong cột Discount sau khi xử lý:")
print(df['Discount'].isnull().sum())
```

##Tiến hành xử lý giá trị thiếu ở cột CustomerID : Thay thế tất cả các giá trị bị thiếu bằng chuỗi ký tự 'GUEST' .

```
# Thay thế giá trị thiếu bằng 'GUEST'
df['CustomerID'] = df['CustomerID'].fillna('GUEST')

# Kiểm tra kết quả
print(df['CustomerID'].head())
print("\nSố giá trị thiếu còn lại:", df['CustomerID'].isnull().sum())
```

# Kiểm tra lại để đảm bảo tập dữ liệu không còn giá trị bị thiếu nào.

```
# Tổng số giá trị thiếu trong toàn bộ tập dữ liệu
print("Tổng số giá trị thiếu:", df.isnull().sum().sum())
```

## Chuyển đổi kiểu dữ liệu của cột Date từ chuỗi (object) sang kiểu dữ liệu thời gian



```
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'].dtype)
```

## Kiểm tra xem có dữ liệu nào bị trùng lặp toàn bộ thông tin trên các cột không. Nếu có, hãy tiến hành loại bỏ các dòng trùng lặp đó.


```
# Đếm số dòng trùng lặp
so_dong_trung_lap = df.duplicated().sum()
print("Số dòng trùng lặp:", so_dong_trung_lap)
# Loại bỏ các dòng trùng lặp
df.drop_duplicates(inplace=True)
# Kiểm tra lại
print("Số dòng trùng lặp sau khi xóa:", df.duplicated().sum())
```

##Phần 3: Trích xuất và Biến đổi dữ liệu (Feature Engineering)

# Tạo một cột mới tên là Revenue (Doanh thu). Biết công thức tính là:
# Revenue = Quantity * UnitPrice * (1 - Discount) .

```
df['Revenue'] = df['Quantity'] * df['UnitPrice'] * (1 - df['Discount'])
print(df[['Quantity', 'UnitPrice', 'Discount', 'Revenue']].head())
```

#Trích xuất Năm (Year) từ cột Date và lưu vào một cột mới tên là Year .


```
# Tạo cột Year từ cột Date
df['Year'] = df['Date'].dt.year

# Kiểm tra kết quả
print(df[['Date', 'Year']].head())
```

## Trích xuất Tháng (Month) từ cột Date và lưu vào một cột mới tên là Month .

```
# Trích xuất tháng từ cột Date
df['Month'] = df['Date'].dt.month

# Hiển thị kết quả
print(df[['Date', 'Month']].head())
```

##Trích xuất Ngày trong tuần (Day of Week) từ cột Date và lưu vào một cột mới tên là
##DayOfWeek (Hiển thị dưới dạng số hoặc tên ngày đều được).


```
# Đảm bảo cột Date có kiểu datetime
df['Date'] = pd.to_datetime(df['Date'])

# Trích xuất ngày trong tuần và lưu vào cột DayOfWeek
df['DayOfWeek'] = df['Date'].dt.day_name()

# Hiển thị kết quả
print(df[['Date', 'DayOfWeek']].head())
```

## Tạo một cột phân loại có tên là Price_Segment dựa trên đơn giá ( UnitPrice ) của sản
phẩm:
- 'High' nếu UnitPrice >= 150.
- 'Medium' nếu 50 <= UnitPrice < 150.
- 'Low' nếu UnitPrice < 50

```
# Phân loại sản phẩm theo đơn giá
def classify_price(price):
    if price >= 150:
        return 'High'
    elif price >= 50:
        return 'Medium'
    else:
        return 'Low'

# Tạo cột Price_Segment
df['Price_Segment'] = df['UnitPrice'].apply(classify_price)

# Hiển thị kết quả
print(df[['UnitPrice', 'Price_Segment']].head())
```

## Phần 4: Lọc và Truy vấn dữ liệu

```
# 1. Đơn hàng khu vực North có Revenue > 300
north_orders = df[(df['Region'] == 'North') & (df['Revenue'] > 300)]

print("Đơn hàng khu vực North có Revenue > 300:")
print(north_orders)


# 2. Số đơn hàng Electronics không được giảm giá
electronics_no_discount = df[
    (df['Product_Category'] == 'Electronics') &
    (df['Discount'] == 0)
]

print("\nSố đơn hàng Electronics không được giảm giá:")
print(len(electronics_no_discount))


# 3. Đơn hàng từ tháng 3 đến tháng 6
orders_mar_jun = df[
    (df['Month'] >= 3) &
    (df['Month'] <= 6)
]

print("\nĐơn hàng từ tháng 3 đến tháng 6:")
print(orders_mar_jun)


# 4. Top 10 đơn hàng có doanh thu cao nhất
top10 = df.sort_values(by='Revenue', ascending=False).head(10)

print("\nTop 10 đơn hàng doanh thu cao nhất:")
print(top10[['OrderID', 'Product_Name', 'Revenue']])


# 5. Đơn hàng có số lượng mua lớn nhất
max_quantity_order = df[df['Quantity'] == df['Quantity'].max()]

print("\nĐơn hàng có số lượng mua lớn nhất:")
print(max_quantity_order)
```

## Phần 5: Phân tích, Gom nhóm và Tổng hợp


```
# Nhóm theo Region và tính tổng Revenue
revenue_by_region = df.groupby('Region')['Revenue'].sum()

print("Tổng doanh thu theo khu vực:")
print(revenue_by_region)
```

```
# Nhóm theo Product_Category và tính trung bình
avg_stats = df.groupby('Product_Category')[['Quantity', 'UnitPrice']].mean()

print("Trung bình Quantity và UnitPrice theo từng danh mục sản phẩm:")
print(avg_stats)
```

```
# Đếm số đơn hàng theo tháng
orders_by_month = df.groupby('Month').size().sort_values(ascending=False)

print("Số lượng đơn hàng theo từng tháng:")
print(orders_by_month)

# Tháng có nhiều đơn hàng nhất
top_month = orders_by_month.idxmax()
max_orders = orders_by_month.max()

print(f"Tháng có nhiều đơn hàng nhất: {top_month}")
print(f"Số đơn hàng: {max_orders}")
```

```

top3_customers = (
    df[df['CustomerID'] != 'GUEST']
    .groupby('CustomerID')['Revenue']
    .sum()
    .sort_values(ascending=False)
    .head(3)
)
print("Top 3 khách hàng có tổng chi tiêu cao nhất:")
print(top3_customers)
```

#Phần 6: Trực quan hóa dữ liệu cơ bản

```
import matplotlib.pyplot as plt


revenue_by_category = df.groupby('Product_Category')['Revenue'].sum()

plt.figure(figsize=(8, 5))
revenue_by_category.plot(kind='bar', color='skyblue')

plt.title('Tổng doanh thu theo Danh mục sản phẩm')
plt.xlabel('Danh mục sản phẩm')
plt.ylabel('Tổng doanh thu')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()



revenue_by_month = df.groupby('Month')['Revenue'].sum().sort_index()

plt.figure(figsize=(8, 5))
plt.plot(
    revenue_by_month.index,
    revenue_by_month.values,
    marker='o',
    linestyle='-',
    color='red'
)

plt.title('Biến động tổng doanh thu theo tháng')
plt.xlabel('Tháng')
plt.ylabel('Tổng doanh thu')
plt.xticks(revenue_by_month.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```
